```python
import re
from io import StringIO
import pandas as pd
import numpy
```

```python
df = pd.read_csv('vendas_2023.csv', encoding='ISO-8859-1')
df
```

```python
#Verificando a causa do problema da linha 3 '2'
with open('vendas_2023.csv', 'r', encoding='ISO-8859-1') as f:
    vendas_tratando = f.read()

vendas_tratando
```

```python
#Tratando as linhas e as possiveis causas invisiveis
vendas_tratando = re.sub(r'[\t\u200b\u200c\u200d]', '', vendas_tratando)
vendas_tratando = vendas_tratando.replace('""', '"')
vendas_tratando = re.sub(r'^"(.*)"$', r'\1', vendas_tratando, flags=re.MULTILINE)

vendas_tratando
```

```python
#lendo o arquivo que havia tratado
df = pd.read_csv(
    StringIO(vendas_tratando),
    sep=',',
    decimal=',',
    quotechar='"',
    engine='python'
)


df.head()
```

```python
#Verificando as informações do arquivo
df.info()
```

```python
#renomeando e padronizando o cabeçalho

df = df.rename(columns={"ID":"id",
               "Data":"data",
               "Produto":"produto",
               "Categoria":"categoria",
               "Quantidade":"quantidade",
               "Preço":"preco"})
```

```python
#Verificando missing value

df.isnull().sum()
```

```python
#Removendo missing value dos Id, não deve haver ID nulo
df = df.dropna(subset=['id'])
```

```python
df.isnull().sum()
```

```python
#Visualizando resultado
df
```

```python
#Removendo produto que não possui registro
df = df.dropna(subset=['produto'])
```

```python
#Removendo Duplicatas do Id. Não deve haver ID duplicado

df = df.drop_duplicates(['id'])
df.duplicated().sum()
```

```python
#Verificando resultado
df
```

```python
#Padronizando os valores

df['preco'] = (df['preco']
               .astype(str)
               .str.replace(r'R\$\s?', '', regex=True)
               .str.replace(',', '.')                 
               .astype(float))

df
```

```python
#Substituindo extenso de 0 a 9 por numeros

df['quantidade'] = df['quantidade'].replace({
    "zero": 0,
    "um": 1,
    "dois": 2,
    "três": 3,
    "quatro": 4,
    "cinco": 5,
    "seis": 6,
    "sete": 7,
    "oito": 8,
    "nove": 9
})
```

```python
#Verificando resultado
df
```

```python
#Convertendo ID que estava float por int
df = df.astype({'id':'int'})

df.info()
```

```python
#Verificando resultado
df
```

```python
#Exportando em 2 arquivos, .DB para utilizar no SQL e .CSV

df.to_sql('data_clean', con='sqlite:///data_clean.db', if_exists='replace', index=False)

df.to_csv("data_clean.csv", index=False, encoding='utf-8-sig')
```
