```python
import re
from io import StringIO
import pandas as pd

```

```python
df = pd.read_csv('vendas_2023.csv', encoding='ISO-8859-1')
df
```

```python
with open('vendas_2023.csv', 'r', encoding='ISO-8859-1') as f:
    vendas_tratando = f.read()

vendas_tratando
```

```python
vendas_tratando = re.sub(r'[\t\u200b\u200c\u200d]', '', vendas_tratando)
vendas_tratando = vendas_tratando.replace('""', '"')
vendas_tratando = re.sub(r'^"(.*)"$', r'\1', vendas_tratando, flags=re.MULTILINE)

vendas_tratando
```

```python
df = pd.read_csv(
    StringIO(vendas_tratando),
    sep=',',
    decimal=',',
    quotechar='"',
    engine='python'
)


df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={"ID":"id",
               "Data":"data",
               "Produto":"produto",
               "Categoria":"categoria",
               "preço":"preco"})
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['id'])
```
