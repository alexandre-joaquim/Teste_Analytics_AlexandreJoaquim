```python
import pandas as pd
```

```python
df = pd.read_csv('data_clean.csv')
df
```

```python
df['valor_total_linha'] = df['quantidade'] * df['preco'] #Criando a coluna de quantidade*preço (total da compra)
```

```python
df['preco'] = pd.to_numeric(df['preco']) # Garantindo que o preço será numero

total_por_produto = df.groupby('produto')['valor_total_linha'].sum().reset_index() #Agrupando e somando os totais da compra

total_por_produto = total_por_produto.sort_values(by='valor_total_linha', ascending=False)

print("total de vendas por produto: ")
print(total_por_produto)

print("\n======================================================================")

top_produto = total_por_produto.loc[total_por_produto['valor_total_linha'].idxmax()] # Localizando o produto com maior valor

print(f"\nO produto com maior faturamento foi: {top_produto['produto']}")
print(f"Valor Total: R$ {top_produto['valor_total_linha']:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")) # Padronizando o valor 
```

```python
total_por_categoria = df.groupby('categoria')['valor_total_linha'].sum().reset_index() #Agrupando e somando os totais da compra

total_por_categoria = total_por_categoria.sort_values(by='valor_total_linha', ascending=False)

print("total de vendas por categoria: ")
print(total_por_categoria)

print("\n======================================================================")

top_categoria = total_por_categoria.loc[total_por_categoria['valor_total_linha'].idxmax()] # Localizando o produto com maior valor

print(f"\nO categoria com maior faturamento foi: {top_categoria['categoria']}")
print(f"Valor Total: R$ {top_categoria['valor_total_linha']:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")) # Padronizando o valor 
```
